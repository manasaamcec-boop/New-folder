```python
import pandas as pd

# Step 1: Load your dataset
df = pd.read_csv("wilsons_data_no_name.csv")

print("Before conversion:")
print(df.head())

# Step 2: Convert all numeric (float) columns to integer
for col in df.select_dtypes(include=['float']).columns:
    df[col] = df[col].astype(int)

# Step 3: Show after conversion
print("\nAfter conversion:")
print(df.head())

# Step 4: Save the updated dataset
df.to_csv("wilsons_data_int.csv", index=False)

print("\n✅ All float values converted to integers!")
print("✨ New dataset saved as 'wilsons_data_int.csv'")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
lr_accuracy = r2_score(y_test,y_predict)*100
```
